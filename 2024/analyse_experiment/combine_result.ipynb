### 1. Select dataset type

```python
import pandas as pd

file_path = 'ExperimentRanV2.xlsx'

df_summary = pd.read_excel(file_path, sheet_name='SelfDatasetSummary')
DATASET_TYPES = list(df_summary['Type'].unique())
```

```python
AUGMENT_METHODS = ["magnitude_warp","window_warping","fourier_transform", "short_time_fourier_transform"]
```

```python
AUGMENT_METHOD = AUGMENT_METHODS[3]
```

```python
for dataset_type in DATASET_TYPES:
    query_result = df_summary.query(f"Type == '{dataset_type}'")
    seleted_dataset = query_result['Name'].to_list()

    df = pd.read_excel(file_path)

    new_df = df

    for i in range(df.columns.get_loc('progress')+1, len(df.columns)):
        if not (df.columns[i] in seleted_dataset):
            new_df = new_df.drop(columns=[df.columns[i]], inplace=False)

    query_result = new_df.query(f"augmentation == '{AUGMENT_METHOD}' or augmentation == 'baseline'")
    query_result = query_result.query("concat_original != True")

    df = query_result


    # Convert the numeric columns to float, excluding non-numeric columns
    numeric_columns = df.columns.drop(['no', 'model', 'augmentation', 'param1_name', 'param1_value', 'param2_name', 'param2_value', 'concat_original', 'progress'])
    df[numeric_columns] = df[numeric_columns].astype(float)

    # Calculate mean excluding the baseline row
    mean_excluding_baseline = df[df['augmentation'] != 'baseline'].mean()

    # Calculate 25th percentile excluding the baseline row
    percentile_25_excluding_baseline = df[df['augmentation'] != 'baseline'].quantile(0.25)

    # Calculate 75th percentile excluding the baseline row
    percentile_75_excluding_baseline = df[df['augmentation'] != 'baseline'].quantile(0.75)

    # Calculate how many rows have values greater than the baseline row
    rows_greater_than_baseline = (df[numeric_columns] > df.loc[0, numeric_columns]).sum()


    # Calculate how many rows have values greater than the baseline row
    rows_equal_than_baseline = (df[numeric_columns] == df.loc[0, numeric_columns]).sum()

    # Calculate how many rows have values less than the baseline row
    rows_less_than_baseline = (df[numeric_columns] < df.loc[0, numeric_columns]).sum()

    # Combine results into a DataFrame
    results = pd.DataFrame({
        'Mean Excluding Baseline': mean_excluding_baseline,
        '25th Percentile Excluding Baseline': percentile_25_excluding_baseline,
        '75th Percentile Excluding Baseline': percentile_75_excluding_baseline,
        'Rows Greater Than Baseline': rows_greater_than_baseline,
        'Row Equal to Baseline': rows_equal_than_baseline,
        'Rows Less Than Baseline': rows_less_than_baseline
    })

    transposed_result = results.transpose()

    concatenated_df = pd.concat([query_result, transposed_result])

    # Open the Excel file in append mode if it exists
    with pd.ExcelWriter('output.xlsx', mode='a', engine='openpyxl') as writer:
        # Write the original DataFrame to the Excel file
        concatenated_df.to_excel(writer, sheet_name=f'{AUGMENT_METHOD}_{dataset_type}', index=False)
```
